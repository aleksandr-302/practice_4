```python
start_number=int(input('стартовое количество: '))
increase=float(input('среднесуточное увеличение в процентах: '))
days=int(input('количество дней для размножения: '))
print('День   Популяция')
population=0
for i in range(0,days):
  population=start_number*(1+increase/100)**i
  print(i+1,'     ',population)
```
